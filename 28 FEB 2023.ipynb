# Multiple Linear Regression

### Loading the dependencies

```python
import pandas as pd
import numpy as np
import klib as k
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading dataset

```python
df = pd.read_csv("advertising.csv")
df
```

## Correlation

```python
k.corr_plot(data=df,target='Sales')
```

```python
sns.heatmap(df.corr(),annot=False)
plt.show
```

# Selecting the features, here only one column

```python
x = df[['TV','Radio']].values
x
```

## Selecting the target column 

```python
y = df[['Sales']].values
```

```python
sc = StandardScaler()
```

```python
x =sc.fit_transform(x)
x
```

```python
y
```

# Splitting into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0);
```

# model building

```python
model = LinearRegression()
```

# fitting the model

```python
model.fit(x_train,y_train)
```

# coefficient

```python
model.coef_
```

# intercept

```python
model.intercept_
```

# training accuracy

```python
training_score = model.score(x_train, y_train)*100
training_score
```

# testing acuracy

```python
testing_score = model.score(x_test, y_test)*100
testing_score
```

# R2 score

```python
R2_score = r2_score(y_test, model.predict(x_test))*100
R2_score
```

# Mean absolute error

```python
mae = mean_absolute_error(y_test, model.predict(x_test))
mae
```

```python
mse = mean_squared_error(y_test, model.predict(x_test))
mse
```

# predicting values

```python
y_pred = model.predict(x)
y_pred
```

# creating a DataFrame with actual value and predicted value inorder nto compare them

```python
OUTPUT = pd.DataFrame(zip(y,y_pred), columns = ("ACTUAL","PREDICTED"), dtype = float)
OUTPUT.head()
```

# Visulising the output

```python
plt.figure(figsize=(12,6))
plt.title("PREDICTED VS ACTUAL DATA")
plt.xticks()
sns.scatterplot(data=OUTPUT, x="ACTUAL", y = "PREDICTED");
```

```python
plt.figure(figsize=(12,6))
plt.title("PRICTED VS ACTUAL DATA")
sns.lineplot(data=OUTPUT, x="ACTUAL", y = "PREDICTED");
```

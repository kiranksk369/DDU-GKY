```python
df.dropna(inplace=True)
df.info()
```

```python
df.Latest_Launch.unique()
```

```python
df['Latest_Launch'] = pd.to_datetime(df.Latest_Launch)
```

## check the dtype of it: 

```python
df.info()
```

## we can see that it have changed to datetime 64[ns] format

## Extracting year, month, date from the latest_launch column

```python
df['Year'] = pd.DatetimeIndex(df.Latest_Launch).year
```

```python
df['month'] = pd.DatetimeIndex(df.Latest_Launch).month
```

```python
df['day'] = pd.DatetimeIndex(df.Latest_Launch).day
```

```python
df.head(10)
```

## lets convent the data type of fuel efficiency into integer 

```python
df.Fuel_efficiency = df.Fuel_efficiency.astype(int)
df.info()
```
